```python
import pandas as pd

excel_file = "Samplefile.xls"
#sheet_name = "sheet1"

df = pd.read_excel(excel_file, header=2, index=False)
```

```python
df_input = df.truncate(before=0, after=3)
df_input['Division']=df_input['Division'].str.cat(sep=' ')
df_input['Type']=df_input['Type'].str.cat(sep=' ')
df_input.index = df_input.index.rename("Index")
df_input
```

```python
#df_input = df_input.drop(['Division', 'Type'], axis=1)
df_output = df_input.set_index(['Type', 'Division', 'Quarter']).T
#df_output = df_output/df_output.sum(axis = 0, skipna = True) * 100
df_output
```

```python
writer = pd.ExcelWriter('output.xlsx')
df_output.to_excel(writer,'Sheet1')
writer.save()
```
